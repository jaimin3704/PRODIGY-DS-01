```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the main data file
data_file_path = 'main.csv'
data = pd.read_csv(data_file_path, skiprows=4)
```

```python
# Select the data for the year 2022
year = '2022'
population_2022 = data[['Country Name', year]].dropna()
```

```python
population_2022
```

```python
population_2022.shape
```

```python
population_2022.isnull().sum()
```

```python
# Histogram for Population Distribution in 2022
plt.figure(figsize=(12, 6))
sns.histplot(population_2022[year], bins=30, kde=True)
plt.title('Distribution of Population in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart for Top 10 Countries by Population in 2022
top_10_countries = population_2022.nlargest(10, year)
plt.figure(figsize=(12, 6))
sns.barplot(x=year, y='Country Name', data=top_10_countries)
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
